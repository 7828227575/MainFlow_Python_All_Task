```python
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
# Read a CSV file
df = pd.read_csv(r'C:\Users\Downloads\Data Cleaning and Preprocessing.csv')

# View the first few rows of the DataFrame
print(df.head())

# Get the shape of the DataFrame (number of rows and columns)
print(df.shape)
#plt.figure(figsize = (25,25))
sns.heatmap(df.isnull())

```

```python
null_var = df.isnull().sum()/df.shape[0]*100
null_var
```

```python
drop_columns = null_var[null_var >10].keys()
drop_columns
```

```python
df2_drop_clm = df.drop(columns=drop_columns)
```

```python
# Check for missing values in the DataFrame
print(df.isnull().sum())

# Drop rows with missing values
df.dropna(inplace=True)

# Fill missing values with a specific value
df.fillna(0, inplace=True)

# Interpolate missing values
df.interpolate(method='linear', inplace=True)
```

```python
# Check for duplicate rows
print(df.duplicated().sum())

# Drop duplicate rows
df.drop_duplicates(inplace=True)
```

```python
# Get the column names
print(df.columns)

# Select a specific column
print(df['ChipMoisture-4 '])

# Filter rows where a condition is true
print(df[df['SteamFlow-4 '] > 70])

# Sort the DataFrame by a column
print(df.sort_values(by='SteamFlow-4 '))

# Group the DataFrame by a column and calculate a summary statistic
print(df.groupby('SteamFlow-4 ').sum())
```

```python
import requests
from bs4 import BeautifulSoup

# Send an HTTP request to the website
url = "https://www.mainflow.in"
response = requests.get(url)

# Parse the HTML content using BeautifulSoup
soup = BeautifulSoup(response.content, 'html.parser')

# Extract the title of the webpage
title = soup.find('title').text
print("Title:", title)

# Extract the paragraphs from the webpage
paragraphs = soup.find_all('p')
for paragraph in paragraphs:
    print("Paragraph:", paragraph.text)
```

```python
import requests
from bs4 import BeautifulSoup

# Send an HTTP request to the website
url = "https://www.mainflow.in"
response = requests.get(url)

# Parse the HTML content using BeautifulSoup
soup = BeautifulSoup(response.content, 'html.parser')

# Extract the title of the webpage
title = soup.find('title').text
print("Title:", title)

# Extract the links from the webpage
links = soup.find_all('a')
for link in links:
    print("Link:", link.get('href'))

# Extract the images from the webpage
images = soup.find_all('img')
for image in images:
    print("Image:", image.get('src'))
```
